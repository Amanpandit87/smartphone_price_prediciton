```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
y_pred = lr.predict(x_test)
```

```python
print("model Prediction = ",r2_score(y_test,y_pred))
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
print("Decision_tree = ",r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error
```

```python
param_dist = {
    'n_estimators': [50, 100, 200],  # Number of estimators in BaggingRegressor
    'estimator__n_estimators': [50, 100, 200],  # Trees in RandomForestRegressor
    'estimator__max_depth': [None, 10, 20, 30],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4],
    'max_samples': [0.5, 0.7, 1.0],
    'max_features': [0.5, 0.7, 1.0],
    'bootstrap': [True, False],
}
```

```python
bag = BaggingRegressor(estimator = rd,n_estimators = 100,max_samples =1.0, max_features= 1.0,bootstrap = False)
```

```python
random_search = RandomizedSearchCV(
    estimator=bag,
    param_distributions=param_dist,
    n_iter=20,  # Number of random combinations to try
    cv=5,       # 5-fold cross-validation
    scoring='neg_mean_squared_error',
    n_jobs=-1,  # Use all cores
    verbose=2,
    random_state=42
)
```

```python
random_search.fit(x_train,y_train)
```

```python
print("Best Parameters:", random_search.best_params_)
print("Best CV Score (MSE):", -random_search.best_score_)
```
